```python
import pandas as pd 
import numpy as np
import matplotlib as mlt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('indian_liver_patient.csv')
df.head()
```

```python
# Корреляция
df.corr()
```

```python
# Вывод составного графика
sns.pairplot(df, hue='Dataset', corner = True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.loc[:, df.columns != 'Dataset'], 
                                                    df['Dataset'],
                                                    random_state=0)

print(f'X_train shape: {x_train.shape}, y_train shape: {y_train.shape},\n'
      f'X_test shape: {x_test.shape}, y_test shape: {y_test.shape}')
```

```python
x_train_concat = np.concatenate((x_train, y_train(437, 1)), axis=1)
x_test_concat = np.concatenate((x_test, y_test(146, 1)), axis=1)
print(f'X_train shape: {x_train_concat.shape},\n'
      f'X_test shape: {x_test_concat.shape}')
```
